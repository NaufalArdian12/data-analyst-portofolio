```python
import pandas as pd

# Baca data
df = pd.read_csv('/content/drive/MyDrive/Data Muaining/data_penjualan_toko_kue.csv')

# Tampilkan 5 baris pertama
df.head()
```

```python
# Cek info kolom dan tipe data
df.info()
```

```python
# Ubah kolom 'tanggal' menjadi datetime
df['tanggal'] = pd.to_datetime(df['tanggal'])


```

```python
# Cek apakah ada missing value
df.isnull().sum()
```

```python
df['produk'].unique()
```

```python
penjualan_harian = df.groupby('tanggal')['total'].sum().reset_index()
penjualan_harian
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))
plt.plot(penjualan_harian['tanggal'], penjualan_harian['total'], marker='o', color='green')
plt.title('Tren Penjualan Harian')
plt.xlabel('Tanggal')
plt.ylabel('Total Penjualan (Rp)')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
produk_terlaris = df.groupby('produk')['jumlah'].sum().sort_values(ascending=False)
produk_terlaris
```

```python
produk_terlaris.plot(kind='bar', color='skyblue')
plt.title('Produk Paling Banyak Terjual')
plt.xlabel('Produk')
plt.ylabel('Jumlah Terjual')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```
